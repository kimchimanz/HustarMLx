## 예측하기

---

예측을 하기 위해 필요한 library와 사용했던 placeholder형식을 그대로 불러옵니다.

```python
import tensorflow as tf
import numpy as np
import pandas as pd

X = tf.placeholder(tf.float32, shape=[None, 4])
Y = tf.placeholder(tf.float32, shape=[None, 1])
W = tf.Variable(tf.random_normal([4, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

가설식을 그대로 가져오고 모델을 초기화해줍니다.

```python
hypothesis = tf.matmul(X, W) + b

saver = tf.train.Saver()
model = tf.global_variables_initializer()
```

이제 아까 저장해두었던 예측할 데이터의 변수를 항목에 넣어주겠습니다.

```python
data_image = pd.read_csv("data image.csv", index_col=0)
data_image.head()
```

```python
year = float(input('연도: '))
pri_ind = float(input('1차산업화: '))
urb_rate = float(input('도시화율: '))
hdi = float(input('HDI: '))
```

어떤 결과값이 나오는지 확인해보겠습니다.

```python
with tf.Session() as sess:
    sess.run(model)
    
    save_path = "./saved.train1.cpkt"
    saver.restore(sess, save_path)
    
    data = ((year, pri_ind, urb_rate, hdi), )
    arr = np.array(data, dtype=np.float32)
    
    x_data = arr[0:4]
    dict = sess.run(hypothesis, feed_dict={X: x_data})
    print("\n%d년에 예상되는 1인당 GDP순위는 %d위입니다."% (year, int(dict[0])))
```

<br>
비용함수가 작은 편이 아니라 어느 정도 예측에 차이가 있습니다. 하지만 관련변수들로 대략적인 1인당GDP순위를 예측할 수 있다는 것을 알았습니다.
그러면 저희가 계산한 미래의 중국의 1차산업비중, 도시화율, 인간개발지수(HDI)를 가지고 미래의 중국의 1인당GDP순위를 예측해 보겠습니다. 

중국 데이터만 모은 파일을 불러오겠습니다.

마지막의 4개의 데이터는 2025년부터 2040년까지 저희가 미래의 중국의 1차산업비중, 도시화율, 인간개발지수(HDI)를 계산하여 예측한 것입니다. 그 연도에 해당하는 GDP 순위를 보면 값이 없는 걸 볼 수 있습니다. 그럼 이 데이터를 가지고 미래의 중국의 1인당 GDP순위를 예측해보겠습니다.

```python
predict = [0, 0, 0, 0, 0]
for i in range(1, 5):
    year = float(input('연도: '% i))
    pri_ind = float(input('1차산업화%d: '% i))
    urb_rate = float(input('도시화율%d: '% i))
    hdi = float(input('HDI%d: '% i))
    
    with tf.Session() as sess:
        sess.run(model)

        save_path = "./saved.train1.cpkt"
        saver.restore(sess, save_path)

        data = ((year, pri_ind, urb_rate, hdi), )
        arr = np.array(data, dtype=np.float32)

        x_data = arr[0:4]
        dict = sess.run(hypothesis, feed_dict={X: x_data})
        predict[i] = int(dict[0])
```

```python
for i in range(1, 5):
    print("%d년에 예상되는 중국의 1인당 GDP순위는 %d위입니다.\n"% (2020+5*i, predict[i]))
```
